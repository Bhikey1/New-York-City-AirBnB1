```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('AB_NYC_2019.csv')
df.head()
```

```python
df.info()
```

```python
# Checking for missing values.
df.isnull().sum()
df = df.dropna()
```

```python
# Checking for and dropping duplicates
df = df.drop_duplicates()
df.duplicated().sum()
```

```python
# What can we learn about different neighborhood areas?
areas = df.groupby('neighbourhood_group', as_index=False).agg({'price':'mean', 'reviews_per_month':'mean'})
areas
```

```python
sns.barplot(x='neighbourhood_group', y='price', data=areas)
```

```python
sns.barplot(x='neighbourhood_group', y='reviews_per_month', data=areas)
```

```python
busiest_areas = df.groupby('neighbourhood_group', as_index=False)['host_id'].count()
busiest_areas
```

```python
sns.barplot(y='neighbourhood_group', x='host_id', data=busiest_areas, orient='h')
```
